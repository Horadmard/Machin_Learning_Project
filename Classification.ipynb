```python
# replace 'Male' with 0 and 'Female' with 1
df['Gender'] = df['Gender'].replace({'Male': 0, 'Female': 1})
df['AcademicPerformance'] = df['AcademicPerformance'].replace({'Below average':1, 'Average':2, 'Good':3, 'Excellent':4})
df['TakingNoteInClass'] = df['TakingNoteInClass'].replace({'Yes': 1,'Sometimes':0, 'No':-1})
df['DepressionStatus'] = df['DepressionStatus'].replace({'Yes': 1,'Sometimes':0, 'No':-1})
df['FaceChallangesToCompleteAcademicTask'] = df['FaceChallangesToCompleteAcademicTask'].replace({'Yes': 1,'Sometimes':0, 'No':-1})
df['LikePresentation'] = df['LikePresentation'].replace({'Yes': 1,'No':0})
df['NumberOfFriend'] = df['NumberOfFriend'].replace({'NaN':0})

df.head(10)
```
